# Import necessary libraries and Load the dataset into a data frame

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
data = pd.read_csv("/content/spam.csv", encoding='latin-1')
```

## Change column names as per need

```python
data = data[['v1', 'v2']]
data.columns = ['spam', 'text']
```

## Split the data into training and testing sets


```python
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['spam'], test_size=0.2, random_state=42)
```

## Preprocessing the data using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)
```

## Create a decision tree classifier

```python
clf = DecisionTreeClassifier(random_state=42)
```

## Train the classifier on the training data

```python
clf.fit(X_train_counts, y_train)
```

## Predict the labels of the test set

```python
y_pred = clf.predict(X_test_counts)
y_pred
```

## Visualize the model

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
plot_tree(clf, filled=True, rounded=True, class_names=['ham', 'spam'], feature_names=vectorizer.get_feature_names_out())
plt.show()
```

## Calculate and print the accuracy score of the classifier

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```
